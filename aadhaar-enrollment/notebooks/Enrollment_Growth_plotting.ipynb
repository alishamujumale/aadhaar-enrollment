```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\Tanaya\Downloads\aadhaar-enrollment-main\aadhaar-enrollment-main\aadhaar-enrollment\notebooks\predicted_enrollment_growth.csv")
print(data.shape)
print(data.head())
```

```python
import pandas as pd

# Take 100 random rows (or any number you want)
sample_data = data.sample(n=100, random_state=42)

# Keep only needed columns
sample_data = sample_data[['state', 'district', 'predicted_growth_rate']]

print(sample_data.head())
```

```python
# Assign each district an index
sample_data = sample_data.reset_index()

plt.figure(figsize=(12,6))
plt.scatter(
    sample_data.index,
    sample_data['predicted_growth_rate'],
    color='orange',
    s=100,
    edgecolors='k'
)
plt.xticks(sample_data.index, sample_data['district'], rotation=90)
plt.ylabel('Predicted Enrollment Growth')
plt.title('District-wise Predicted Aadhaar Enrollment Growth')
plt.grid(True)
plt.tight_layout()
plt.show()
```
